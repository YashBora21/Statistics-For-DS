```python
import numpy as np
arr=np.array([2,3,4,6,7,8,9,12,13,16,17,23,25,27,34,37,201])
arr
q1=np.percentile(arr,25)
q2=np.percentile(arr,50)
q3=np.percentile(arr,75)
IQR=q3-q1
IQR
```

```python
UF=q3+(1.5*IQR)
LF=q1-(1.5*IQR)
print(UF,LF)
```

```python
l=[]
for i in arr:
    if(i>=LF and i<=UF):
        l.append(i)

arr2=np.array(l)
```

```python
print(arr)
print(arr2)
```
